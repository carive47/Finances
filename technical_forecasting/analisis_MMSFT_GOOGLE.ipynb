```python
#API datos
import yfinance as yfin

#Manejo datos
import pandas as pd

#Graficos
import matplotlib.pyplot as plt

import numpy as np
```

```python
ticker = ['GOOG', 'MSFT']

dfG = yfin.download(ticker[0], start='2019-01-01', end='2021-01-01')
dfG
```

```python
dfM = yfin.download(ticker[1], start='2019-01-01', end='2021-01-01')
dfM
```

```python
## guardar tabla
  ## dfG.to_csv('BTC.csv')
```

```python
dfG.describe()
```

```python
dfM.describe()
```

```python
# grafico de precios de cierre

plt.figure(figsize=(15,8))
plt.plot(dfM['Close'], color = 'green')
plt.title ('MSFT [Volume]')
plt.xlabel('Fecha')
plt.ylabel('Volumen')
plt.show()
```

```python
#retorno artmetico
  # adición de la columna t-1
dfG['Adj Close t-1']= dfG['Adj Close'].shift(1)
  # Adicion retorno Aritmetico
dfG['Return'] = (dfG['Adj Close'] - dfG['Adj Close t-1']) / dfG['Adj Close t-1']
dfG
```

```python
#retorno logaritmico

dfG['log return'] = np.log(dfG['Adj Close'])-np.log(dfG['Adj Close t-1'])
dfG
```

```python
#grfica de retornor

plt.figure(figsize = ( 15,8))
plt.plot(dfG['log return'], color = 'red')
plt.title('Retorno log de MSFT')
plt.xlabel('fecha')
plt.ylabel('retorno')
plt.show()
```

```python
# volatilidad diaria
vol_d = np.std(dfG['log return'])
vol_d
```

```python
# volatilidad anualizada
vol_a= vol_d*np.sqrt(252)
vol_a
```

```python
from numpy.lib.function_base import average
ticker3='^GSPC'
dfS=yfin.download(ticker3, start ='2015-01-01', end ='2021-01-01')
dfS['Adj close t-1'] = dfS['Adj Close'].shift(1)
dfS['ret']= (dfS['Adj Close']- dfS['Adj close t-1']) / dfS['Adj close t-1']
yar= np.mean(dfS['ret'])*252
yar*100
```
